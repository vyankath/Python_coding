# Handling of missing values

```python
Follwing some functions are used to handle missing values.

1. df.isna()                       >> To check are any missing values in df.
2. df.isnull()                     >> To check are any missing values in df.
3. df.isna().sum()                 >> Wil return sum of all missing values. i.e. no of misssing values.                    
4. df.isna().mean()                >> Wil return mean of all missing values columns.
5. df.fillna():                    >> Used to fill the null values.
    1. df[col_name].mean()         >> Used to fill the null values by mean value. 
    2. df[col_name].median()       >> Used to fill the null values by median value.
    3. df[col_name].mode()[0]      >> Used to fill the null values by mode value.
    4. df[col_name].min()          >> Used to fill the null values by min value.
    5. df[col_name].max()          >> Used to fill the null values by max value.
    6. df[col_name].(static value) >> Used to fill the null values by static value value.
6. df.dropna(thresh=df.shape[1])
7. df.drop()
```
